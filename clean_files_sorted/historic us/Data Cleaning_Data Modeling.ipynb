```python
import pandas as pd
import os
import glob
path = os.getcwd()
csv_files = glob.glob(os.path.join(path, "*.csv"))
print(csv_files)
```

```python
count = 0
df = pd.DataFrame()
for file in sorted(csv_files):
    df_temp = pd.read_csv(file)
    #df_concat = pd.concat([df_concat, df_temp],axis =1)
    
    df = pd.concat([df.reset_index(drop=True), df_temp.reset_index(drop=True)], axis=1)
    #df_concat = pd.concat([df_concat,df_temp], axis = 1)

    
```

```python
import numpy as np
df1 = df1.replace('.0','', regex=True)
df1 = df1.replace(',','', regex = True)
df1 = df1.replace(' ','', regex=True)
df1 = df1.replace('  ','', regex=True)
df1 = df1.replace('',np.nan)
df1 = df1.replace('*',np.nan)
df1 = df1.replace('/', "", regex = True)
df1 = df1.replace('r', "", regex = True)
df1
```
